```python
WIP = range(0,20)
bottleNeckTH = 0.4
minCT = 30
TH = []
CT = []
for i in range(len(WIP)):
    if WIP[i] < bottleNeckTH * minCT:  
        CT.append(minCT)
        TH.append(float(WIP[i]/CT[i]))
    else:
        TH.append(bottleNeckTH)
        CT.append(float(WIP[i]/TH[i]))
```

```python
import matplotlib.pyplot as plt
plt.plot(WIP, TH)
plt.ylim([0, 0.6])
plt.xlabel('WIP')
plt.ylabel('TH')
```
