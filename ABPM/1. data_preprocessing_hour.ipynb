### 2. 先取前後各一個時間平均做為補值，再丟掉有缺失值的樣本（連續多個NAN）

```python
def replace_nan_with_two_average(row, col):
    if pd.isna(row[col]):
        if col > 0 and col < 23:
            other_columns = [col-1, col+1]
        elif col == 23:
            other_columns = [22, 0]
        elif col == 0:
            other_columns = [23, 1]        
        return np.nanmean(row[other_columns])
    else:
        return row[col]
```

```python
df_drop_two_nan = dfByMeasureValue.copy()
df_drop_two_nan = df_drop_two_nan[~df_drop_two_nan.index.isin(df_drop_two_nan.dropna().index)]
```

```python
for column in df_drop_two_nan.columns:
    df_drop_two_nan[column] = df_drop_two_nan.apply(lambda row: replace_nan_with_two_average(row, column), axis=1)
```

```python
df_drop_two_nan = df_drop_two_nan.dropna()
df_drop_two_nan.to_csv("./preprocessed data/two_avg_drop_nan.csv")
df_drop_two_nan # 1401 rows × 24 columns
```

### 3. 先取前後各兩個時間平均做為補值，再丟掉有缺失值的樣本（連續多個NAN）

```python
def replace_nan_with_four_average(row, col):
    if pd.isna(row[col]):
        if col > 1 and col < 22:
            other_columns = [col-2, col-1, col+1, col+2]
        elif col == 0:
            other_columns = [22, 23, 1, 2]
        elif col == 1:
            other_columns = [23, 0, 2, 3]    
        elif col == 22:
            other_columns = [20, 21, 23, 0]
        elif col == 23:
            other_columns = [21, 22, 0, 1]       
        return np.nanmean(row[other_columns])
    else:
        return row[col]
```

```python
df_drop_four_nan = dfByMeasureValue.copy()
df_drop_four_nan = df_drop_four_nan[~df_drop_four_nan.index.isin(df_drop_four_nan.dropna().index)] # 拿掉完整的142筆(training data)
for column in df_drop_four_nan.columns:
    df_drop_four_nan[column] = df_drop_four_nan.apply(lambda row: replace_nan_with_four_average(row, column), axis=1)
```

```python
df_drop_four_nan = df_drop_four_nan.dropna()
df_drop_four_nan = df_drop_four_nan.round(3)
df_drop_four_nan.to_csv("./preprocessed data/four_avg_drop_nan.csv")
df_drop_four_nan # 1428 rows × 24 columns
```

### 4. 先取前一個&後兩個時間平均做為補值，再丟掉有缺失值的樣本（連續多個NAN）

```python
def replace_nan_with_three_average(row, col, more):
    if more == "next":
        if pd.isna(row[col]):
            if col > 0 and col < 22:
                other_columns = [col-1, col+1, col+2]
            elif col == 0:
                other_columns = [23, 1, 2]
            elif col == 22:
                other_columns = [21, 23, 0]
            elif col == 23:
                other_columns = [22, 0, 1]       
            return np.nanmean(row[other_columns])
        else:
            return row[col]
    else:
        if pd.isna(row[col]):
            if col > 1 and col < 23:
                other_columns = [col-2, col-1, col+1]
            elif col == 0:
                other_columns = [22, 23, 1]
            elif col == 1:
                other_columns = [23, 0, 2]
            elif col == 23:
                other_columns = [21, 22, 0]       
            return np.nanmean(row[other_columns])
        else:
            return row[col]
```

```python
df_drop_three_nan = dfByMeasureValue.copy()
df_drop_three_nan = df_drop_three_nan[~df_drop_three_nan.index.isin(df_drop_three_nan.dropna().index)] # 拿掉完整的142筆(training data)
for column in df_drop_three_nan.columns:
    df_drop_three_nan[column] = df_drop_three_nan.apply(lambda row: replace_nan_with_three_average(row, column, "next"), axis=1)
```

```python
df_drop_three_nan = df_drop_three_nan.dropna()
df_drop_three_nan = df_drop_three_nan.round(3)
df_drop_three_nan.to_csv("./preprocessed data/three_next_avg_drop_nan.csv")
df_drop_three_nan # 1404 rows × 24 columns
```

```python
df_drop_three_nan = dfByMeasureValue.copy()
df_drop_three_nan = df_drop_three_nan[~df_drop_three_nan.index.isin(df_drop_three_nan.dropna().index)] # 拿掉完整的142筆(training data)
for column in df_drop_three_nan.columns:
    df_drop_three_nan[column] = df_drop_three_nan.apply(lambda row: replace_nan_with_three_average(row, column, "pre"), axis=1)
```

```python
df_drop_three_nan = df_drop_three_nan.dropna()
df_drop_three_nan = df_drop_three_nan.round(3)
df_drop_three_nan.to_csv("./preprocessed data/three_pre_avg_drop_nan.csv")
df_drop_three_nan # 1404 rows × 24 columns
```

## 5. 取前後兩個時間bp加權平均

```python
def replace_nan_with_two_weighted(row, col):
    if pd.isna(row[col]):
        if col > 0 and col < 23:
            other_columns = [col-1, col+1]
        elif col == 23:
            other_columns = [22, 0]
        elif col == 0:
            other_columns = [23, 1] 
        weighted_avg = row[other_columns[1]] - (row[other_columns[1]]-row[other_columns[0]])*(row[other_columns[0]]/(row[other_columns[1]]+row[other_columns[0]]))
        return weighted_avg
    else:
        return row[col]
```

```python
df_drop_two_nan_weighted = dfByMeasureValue.copy()
df_drop_two_nan_weighted = df_drop_two_nan_weighted[~df_drop_two_nan_weighted.index.isin(df_drop_two_nan_weighted.dropna().index)] # 拿掉完整的142筆(training data)
for column in df_drop_two_nan_weighted.columns:
    df_drop_two_nan_weighted[column] = df_drop_two_nan_weighted.apply(lambda row: replace_nan_with_two_weighted(row, column), axis=1)
```

```python
df_drop_two_nan_weighted = df_drop_two_nan_weighted.dropna()
df_drop_two_nan_weighted = df_drop_two_nan_weighted.round(3)
df_drop_two_nan_weighted.to_csv("./preprocessed data/two_weighted_drop_nan.csv")
df_drop_two_nan_weighted # 1428 rows × 24 columns
```
